```python
import ssl
import urllib.request
from pubchempy import get_compounds

# Disable SSL certificate verification globally
ssl._create_default_https_context = ssl._create_unverified_context

# Get the SMILES for Chlorpromazine using its name
compounds = get_compounds('Chlorpromazine', 'name')
chlorpromazine_smiles = compounds[0].isomeric_smiles

print(f'The SMILES for Chlorpromazine is: {chlorpromazine_smiles}')
```

```python
from rdkit import Chem
from rdkit.Chem import Draw

# Convert SMILES to a molecular structure
mol = Chem.MolFromSmiles('CN(C)CCCN1C2=CC=CC=C2SC3=C1C=C(C=C3)Cl')

# Display the molecular structure
Draw.MolToImage(mol)
```

```python
from rdkit import Chem
from rdkit.Chem import Descriptors

# Convert SMILES to a molecule
mol = Chem.MolFromSmiles('CN(C)CCCN1C2=CC=CC=C2SC3=C1C=C(C=C3)Cl')

# Calculate molecular weight and logP
mol_weight = Descriptors.MolWt(mol)
logP = Descriptors.MolLogP(mol)

print(f"Molecular Weight: {mol_weight}")
print(f"logP: {logP}")
```
